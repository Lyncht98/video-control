```python
import numpy as np
import cv2 as cv
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Read in video
cap = cv.VideoCapture('video_clips/output_n_0.avi')


# save all frames in a list
frames = list()
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    frames.append(gray)

cap.release()
```
